# 1. Import Necessary libraries¶

```python
import pandas as pd
import numpy as np
```

# 2. Import Data

```python
books_df = pd.read_excel("book_data.xlsx")
books_df
```

# 3. Data Understanding

```python
len(books_df)
```

```python
books_df.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
books_df.head(10)
```

# 4.Feature Engineering

```python
books_df.rename(columns = {'User.ID':'id','Book.Title':'booktitle','Book.Rating':'RATING'}, inplace = True)
books_df.head()
```

```python
#number of unique users in the dataset
len(books_df.id.unique())
```

```python
len(books_df.booktitle.unique())
```

# creating pivot table

```python
user_books_df = books_df.pivot_table(index='id',
                                 columns='booktitle',
                                 values='RATING')
user_books_df
```

```python
user_books_df.index = books_df.id.unique()
```

```python
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
```

```python
user_books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
```

```python
user_sim
```

### other way to get cosine correlation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x=cosine_similarity(user_books_df.values)
```

```python
pdx=pd.DataFrame(x)
pdx
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```
